```python
import numpy as np 
import pandas as pd
match = pd.read_csv('matches.csv')
delevery = pd.read_csv('deliveries.csv')
match
```

```python
total_score_df = delevery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score_df
```

```python
total_score_df = total_score_df[total_score_df['inning'] == 1]
```

```python
total_score_df
```

```python
match_df  = match.merge(total_score_df[["match_id",'total_runs']],left_on = "id" , right_on="match_id")
```

```python
match_df
```

```python
match_df['team1'].unique()
```

```python
teams = {
    'Sunrisers Hyderabad',
    'Mumbai Indians', 
    'Gujarat Lions',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders', 
    'Kings XI Punjab', 
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
    
}
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
match_df
```

```python
match_df = match_df[['match_id' , 'city' , 'winner' , 'total_runs']]
```

```python
delevery_df = match_df.merge(delevery, on = 'match_id')
```

```python
delevery_df  = delevery_df[delevery_df['inning'] == 2]
```

```python
delevery_df.shape
```

```python
delevery_df['current_score'] = delevery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delevery_df
```

```python
delevery_df['runs_left'] = delevery_df['total_runs_x'] - delevery_df['current_score']
```

```python
delevery_df
```

```python
delevery_df['balls_left'] = 126 - (delevery_df['over']*6 + delevery_df['ball'])
```

```python
delevery_df
```

```python
delevery_df['player_dismissed'] = delevery_df['player_dismissed'].fillna("0")
delevery_df['player_dismissed'] = delevery_df['player_dismissed'].apply(lambda x: "1" if x != "0" else "0")

delevery_df['player_dismissed'] = delevery_df['player_dismissed'].astype('int')
wickets = delevery_df.groupby('match_id').cumsum()['player_dismissed'].values
delevery_df['wickets'] = 10 - wickets
delevery_df.head()
```

```python
 delevery_df['crr'] = (delevery_df['current_score'] * 6) / (120 - delevery_df['balls_left'])
```

```python
 delevery_df['rrr'] = (delevery_df['runs_left'] * 6) / delevery_df['balls_left']
```

```python
 delevery_df
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
  delevery_df['result'] = delevery_df.apply(result,axis = 1)
```

```python
 delevery_df
```

```python
 final_df  = delevery_df[['batting_team' , 'bowling_team' , 'city' , 'runs_left' , 'balls_left' , 'wickets' , 'total_runs_x', 'crr' ,'rrr' , 'result']]
    
```

```python
x  =  final_df.iloc[:,: -1]
y =  final_df.iloc[: , -1]
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2, random_state =1 )
```

```python
x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf  = ColumnTransformer([
    ('trf' , OneHotEncoder(sparse = False,drop = 'first'),['batting_team','bowling_team' ,'city'])
] , remainder = 'passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe  = Pipeline(steps=[
    ('steps' ,trf),
    ('step2' ,  LogisticRegression(solver = 'liblinear'))
])
```

```python
pipe.fit(x_train , y_train)
```
